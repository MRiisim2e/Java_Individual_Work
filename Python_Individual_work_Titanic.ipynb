```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = df.dropna(subset=['Age'])

#hereby I decided to drop all NaN values in column 'Age' so I could use only proper data for data visulization
```

```python
df['Age'].value_counts(dropna=False)
```

```python
# CUSTOM BIN EDGES
bin_edges = [0, 10, 20, 30, 40, 50, 60, 70, 80]
plt.hist(df['Age'], bins=bin_edges, color='pink', edgecolor='white')

plt.title('Age distribution among passengers') #we added title
plt.xlabel('Age') #title on x axis
plt.ylabel('Number of passengers') #title on y axis
```
